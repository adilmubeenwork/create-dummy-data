```python
import pandas as pd
import random
from faker import Faker
from random import randrange
from datetime import datetime
```

```python
import faker


nr_of_customers = 1000

fake = Faker('de_DE')

customers = []

for customers_id in range(nr_of_customers):

    # Create start and end date 
    start_date = fake.date()
    end_date = fake.date()
    start_time = fake.time()
    end_time = fake.time()
    time_list = []
    for i in range(10):
     if start_time > end_time:
      start_time = fake.time()
     else:
      time_list.append(str(start_time))
      break
    time_string = ''.join(map(str, time_list))
    

    watering_time = fake.time()

    Plants = fake.numerify()

    #create customer's name
    Name = fake.name()

    
    status = fake.boolean()

    if status is True:
      status='Paid'
    else:
      status='Not Paid'

    #create place name

    Place_name = fake.city_name()

    seed_id = fake.iana_id

    #plant type
    street_address = fake.city_suffix()

    # Create gender
    #gender = random.choice(["M", "F"]))

    #Create city
    city = fake.city()

    #create product ID in 8-digit barcode
    Place_ID = fake.ean(length=8)
    
    #create amount spent
    Zip_code = fake.postcode()
    #(right_digits=2, positive=True, min_value=1, max_value=100)

    #createdseed
  
    seedid = fake.numerify()
    category1 = ['nucleus seed', 'breeder seed', 'foundation seed', 'certified seed']
    category = random.choice(category1)
    seeds1 = ['FlaxSeeds', 'Chia Seeds', 'Rajgira Seeds', 'Sunflower Seeds', 'Pumpkin Seeds', 'Basil Seeds', 'Hemp Seeds', 'Pomegranate Seeds']
    seeds = random.choice(seeds1)
    priceperseed = fake.pyfloat(positive=True, right_digits=2, left_digits=2)


    #factors
    season1 = ['Summer', 'Winter', 'Spring', 'Autumn']
    season = random.choice(season1)

    holidays = fake.day_of_week()
    specialevent = fake.boolean()

    if specialevent is True:
      specialevent='Premium'
    else:
      specialevent='Standard'
    historicw1 = ['Sunny', 'Rainy', 'Windy', 'Hot', 'Cloudy', 'Stormy', 'Cold']
    historicweather = random.choice(historicw1)
    competitorprices = fake.pyfloat(positive=True, right_digits=2, left_digits=2)


    #purchaseplant
    plantid = fake.numerify()
    farmerid = fake.numerify()
    pricepayed = fake.pyfloat(positive=True, right_digits=2, left_digits=2)
    timestamp = fake.date_time()
    plantstatus = fake.boolean()

    if plantstatus is True:
      plantstatus='Active'
    else:
      plantstatus='Cancel'
    pickuptimestamp = fake.date_time()
    partnerid = fake.numerify()
    marketdata = fake.numerify()
    #farmer
    farmerid = fake.numerify()
    firstname1 = Name
    parsed_name1 = HumanName(firstname1)
    firstname = parsed_name1.first
    lastname1 = Name
    parsed_name = HumanName(lastname1)
    lastname = parsed_name.last
    birthdate = fake.date()
    gender = random.choice(["M", "F"])









    customers.append([Place_ID, Place_name, Name, start_date, end_date, time_string, end_time, watering_time, Plants, status, street_address, Zip_code, city, seedid , category, seeds, priceperseed, season, holidays, specialevent, historicweather, competitorprices, plantid, farmerid, pricepayed, timestamp, plantstatus, pickuptimestamp, partnerid, marketdata,  farmerid, firstname, lastname, birthdate, gender])


customers_df = pd.DataFrame(customers, columns=['Place_id', 'Place_name', 'Name', 'start_date', 'end_date', 'start_time', 'end_time', 'watering_time', 'Plants', 'status', 'street_address', 'Zip_code', 'city', 'seedid' , 'category', 'seeds',  'priceperseed', 'season', 'holidays', 'specialevent', 'historicweather', 'competitorprices', 'plantid', 'farmerid', 'pricepayed', 'timestamp', 'plantstatus', 'pickuptimestamp', 'partnerid', 'marketdata',  'farmerid', 'firstname', 'lastname', 'birthdate', 'gender']) 
                
pd.pandas.set_option('display.max_columns', None)
customers_df.head(10)
```
